```python
import pandas as pd
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
recent_income_to_load = Path("2021_non_fam_income.csv")

most_recent = pd.read_csv(recent_income_to_load)
```

```python


# Assuming your DataFrame is named 'most_recent_drop'
filtered_df = most_recent.iloc[:10]

# Remove the percentage symbol and convert 'Nonfamily Households' column to float
filtered_df['Nonfamily Households'] = filtered_df['Nonfamily Households'].str.rstrip('%').astype(float)

# Define the data for the pie chart
values = filtered_df['Nonfamily Households']

# Create the pie chart
plt.figure(figsize=(10, 10))
plt.pie(values, autopct='%1.1f%%', startangle=90)

# Create an associated legend with labels
plt.legend(filtered_df['Label (Grouping)'], loc='best')

# Add a title
plt.title('2021 Income for Individual Households')

# Display the pie chart
plt.show()
```
