```python
import pandas as pd
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
housing_to_load = Path("Resources/LIHTCPUB.CSV")

public_housing_data = pd.read_csv(housing_to_load)
```

```python
public_housing_data.head(1)
```

```python
column_names = public_housing_data.columns.tolist()
print(column_names)
```

```python
#OR only 
updated_housing_data = public_housing_data[public_housing_data['proj_st'] == 'OR']


unique_names = updated_housing_data['proj_st'].unique()
print(unique_names)
```

```python
#Filter only Multnmah Cty using the fips code, 41 = OR  05 = Mult Cty, filter 
multnomah_housing_data = updated_housing_data[updated_housing_data['fips2010'].str.startswith('4105')]
```

```python
#Filter only Gresham and Portland
multnomah_housing_data = updated_housing_data[(updated_housing_data['proj_cty'] == 'PORTLAND') | (updated_housing_data['proj_cty'] == 'GRESHAM')]
```

```python
unique_blocks = multnomah_housing_data['fips2010'].unique()
print(unique_blocks)
```

```python
row_count = len(multnomah_housing_data)
print(row_count)
```
